```python


# Load pre-trained model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-ar'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Ensure the model is on the desired device (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Define your dataset class for bidirectional translation
class BidirectionalTranslationDataset(Dataset):
    def __init__(self, dataframe, tokenizer):
        self.data = dataframe
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        eng_text = self.data['English'][idx]
        arabic_text = self.data['Arabic'][idx]

        # Encode text, ensuring padding and truncation
        eng_tokenized = self.tokenizer.encode(
            eng_text, return_tensors="pt", padding="max_length", truncation=True, max_length=64
        )
        arabic_tokenized = self.tokenizer.encode(
            arabic_text, return_tensors="pt", padding="max_length", truncation=True, max_length=64
        )

        # Return both directions: Arabic to English and English to Arabic
        return {
            "input_ids_arabic_to_english": arabic_tokenized.flatten(),
            "labels_arabic_to_english": eng_tokenized.flatten(),
            "input_ids_english_to_arabic": eng_tokenized.flatten(),
            "labels_english_to_arabic": arabic_tokenized.flatten()
        }

# Custom collate function to handle sequences of varying lengths for bidirectional translation
def bidirectional_collate_fn(batch):
    input_ids_arabic_to_english = [item["input_ids_arabic_to_english"] for item in batch]
    labels_arabic_to_english = [item["labels_arabic_to_english"] for item in batch]
    input_ids_english_to_arabic = [item["input_ids_english_to_arabic"] for item in batch]
    labels_english_to_arabic = [item["labels_english_to_arabic"] for item in batch]

    # Pad sequences to the same length within each batch, considering device
    input_ids_arabic_to_english_padded = torch.nn.utils.rnn.pad_sequence(input_ids_arabic_to_english, batch_first=True).to(device)
    labels_arabic_to_english_padded = torch.nn.utils.rnn.pad_sequence(labels_arabic_to_english, batch_first=True).to(device)
    input_ids_english_to_arabic_padded = torch.nn.utils.rnn.pad_sequence(input_ids_english_to_arabic, batch_first=True).to(device)
    labels_english_to_arabic_padded = torch.nn.utils.rnn.pad_sequence(labels_english_to_arabic, batch_first=True).to(device)

    return {
        "input_ids_arabic_to_english": input_ids_arabic_to_english_padded,
        "labels_arabic_to_english": labels_arabic_to_english_padded,
        "input_ids_english_to_arabic": input_ids_english_to_arabic_padded,
        "labels_english_to_arabic": labels_english_to_arabic_padded
    }

# Create datasets and combine them into a single dataloader with custom collate function
dataset_arabic_to_english = BidirectionalTranslationDataset(DATA, tokenizer)
dataset_english_to_arabic = BidirectionalTranslationDataset(DATA, tokenizer)
combined_dataset = torch.utils.data.ConcatDataset([dataset_arabic_to_english, dataset_english_to_arabic])
dataloader = DataLoader(combined_dataset, batch_size=4, shuffle=True, collate_fn=bidirectional_collate_fn)

# Training loop (simplified)
# Create the optimizer (replace with your desired optimizer)
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)  # Example optimizer
EPOCHS = 10

class EpochProgressCallback:
    def __init__(self, optimizer, dataloader, num_epochs, print_freq=100):
        self.optimizer = optimizer
        self.dataloader = dataloader
        self.num_epochs = num_epochs
        self.print_freq = print_freq
        self.current_step = 0
        self.current_epoch = 0
        self.num_training_steps = len(dataloader) * num_epochs

    def on_batch_end(self, loss):
        self.current_step += 1
        if self.current_step % self.print_freq == 0:
            epoch_progress = self.current_step / self.num_training_steps
            print(f"Epoch: {self.current_epoch + 1}/{self.num_epochs}, Step: {self.current_step}/{self.num_training_steps}, Progress: {epoch_progress:.2f}, Loss: {loss:.4f}")

    def on_epoch_end(self):
        self.current_epoch += 1

# Update callback initialization
callback = EpochProgressCallback(optimizer, dataloader, EPOCHS)


num_training_steps = len(dataloader) * EPOCHS  
# callback = EpochProgressCallback(model, optimizer, num_training_steps)  # Pass both model and optimizer

for epoch in range(EPOCHS):
    for batch in dataloader:
        input_ids_arabic_to_english = batch["input_ids_arabic_to_english"]
        labels_arabic_to_english = batch["labels_arabic_to_english"]
        input_ids_english_to_arabic = batch["input_ids_english_to_arabic"]
        labels_english_to_arabic = batch["labels_english_to_arabic"]

        # Train model for Arabic to English translation
        outputs_arabic_to_english = model(input_ids=input_ids_arabic_to_english, labels=labels_arabic_to_english)
        loss_arabic_to_english = outputs_arabic_to_english.loss

        # Train model for English to Arabic translation
        outputs_english_to_arabic = model(input_ids=input_ids_english_to_arabic, labels=labels_english_to_arabic)
        loss_english_to_arabic = outputs_english_to_arabic.loss

        # Calculate total loss
        total_loss = loss_arabic_to_english + loss_english_to_arabic

        # Backpropagation and optimizer steps
        total_loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        # Print progress
        callback.on_batch_end(total_loss.item())  # Pass the loss as an argument

    # Print progress at the end of each epoch
    callback.on_epoch_end()
```
